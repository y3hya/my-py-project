```python
import openpyxl
import pandas as pd
path = "C:/Dropbox/OpSupport Team Folder/Cell Shop/Commission/AT&T Reports/2021-08/Formatted/Statement.xlsx"
wb = openpyxl.load_workbook(path, data_only=True)
```

```python
for ws in wb.worksheets:
    for tbl in ws._tables:
        if tbl == 'Raw':
            tableRange = ws.tables[tbl].ref.split(":")
            # print(ws.title)
            sheet=wb[ws.title]
```

```python
table = []
for row in sheet[tableRange[0]:tableRange[1]]:
    data_cols = []
    for cell in row:
        data_cols.append(cell.internal_value)
    table.append(data_cols)
```

```python
pd.DataFrame(table)
```
